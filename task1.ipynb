# Load webpages into Python through requests



## Loading page source code into Python

Link: https://en.wikipedia.org/wiki/Ariquemes

```python
import requests as req
```

```python
URL = 'https://en.wikipedia.org/wiki/Ariquemes'
```

```python
r=req.get(URL)
```

```python
print(r.content[:2000])
```

```python
import bs4
```

```python
soup=bs4.BeautifulSoup(r.content)
```

```python
print(soup.prettify())
```
